```python

from sklearn.linear_model import LinearRegression
model=LinearRegression()
from sklearn.model_selection import train_test_split
X=df[["places ","Date"]].values
y=df["value"].values
#dividing dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
#fitting into linear regression model
model.fit(X_train,y_train)


```

```python
model.score(X_train,y_train)
```

```python
predicted=model.predict(X_test)
```

```python
from sklearn .metrics import mean_absolute_error,r2_score
print("r2 score ",r2_score(predicted,y_test))
print("mean_absolute_error",mean_absolute_error(predicted,y_test))
```
